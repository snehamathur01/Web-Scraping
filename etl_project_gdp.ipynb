```python
#Task 1: Extracting information
res=requests.get('https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29')
soup=BeautifulSoup(res.content,"lxml")
table=soup.find('table',{'class':'wikitable'})
countriesall=table.find_all('tr',class_=False)
#print(countriesall)
td_tag=[i.find_all('td') for i in countriesall]
#print(td_tag)
IMF_estimate=[]
for i in td_tag:
    for j in range(len(i)):
        if i[j].has_attr('style'):
            continue
        else:
            IMF_estimate.append(i[j].text.strip(''))
            break
#print(IMF_estimate)
country_names=[c.find("td").find("a") for c in countriesall]
country=[c.text for c in country_names]
#print(country)
df=pd.DataFrame({'Country':country,'GDP_USD_Million':IMF_estimate,'Year':"2023"})
```

```python
#Task 2: Transform information
IMF_Estimate_billion=[i.replace(',','') for i in IMF_estimate]
IMF_Estimate_billion=[round(int(i)/1000,2) for i in IMF_Estimate_billion]
IMF_Estimate_billion
df["GDP_USD_Million"]=IMF_Estimate_billion
df=df.rename(columns={"GDP_USD_Million":"GDP_USD_Billion"})
df
```

```python
df.to_csv("GDP_2023")
```
